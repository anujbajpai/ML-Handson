```python
dataset = pandas.read_csv("D:\datasets\winemag-data_first150k.csv", index_col=0)
dataset.head()
```

```python
#map() returns a new Series where all the values have been transformed by your function.
#map returns series where as apply returns dataframe
#They don't modify the original data they're called on
m = dataset.price.mean()
dataset.price.map(lambda p: p - m)
```

```python
#apply the mean to all rows of dataframe
#map returns series where as apply returns dataframe
#They don't modify the original data they're called on
def meanDiffAllPrice(row):
    row.price = row.price-m
    return row
s = dataset.apply(meanDiffAllPrice, axis = 'columns')
s
```
