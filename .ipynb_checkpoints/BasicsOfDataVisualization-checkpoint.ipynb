```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Samuel Norman "Sam" Seaborn is a fictional character portrayed by Rob Lowe on the television serial drama The West Wing.
import seaborn as sns
#below line is for styling the seaborn plots
pd.plotting.register_matplotlib_converters()
from IPython.display import Image

iris = pd.read_csv('D:\datasets\iris.csv.data', names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

iris.head(4)
```

```python
wine_reviews = pd.read_csv("D:\datasets\winemag-data_first150k.csv", index_col=0)
wine_reviews.head(2)
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

You might like to add a regression line, or the line that best fits the data. We do this by changing the command to sns.regplot.

```python
sns.set_style("dark") #happened becoz of [register_matplotlib_converters]
sns.regplot(x = iris['sepal_length'],y=iris['sepal_width'])
```

```python
#we can use the sns.lmplot command to add regression lines corresponding to each class[hue] of iris with color coded  
sns.lmplot(x='sepal_length', y = 'sepal_width', hue = 'class', data = iris)
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x = iris['sepal_length'], y = iris['petal_length'], hue = iris['class'])
#hue is for colour
```

```python
#we can adapt the design of the scatter plot to feature a categorical variable (like "class") on one of the main axes. We'll refer to this plot type as a categorical scatter plot, and we build it with the sns.swarmplot command.
plt.figure(figsize=(20,6))
sns.swarmplot(x=iris['class'], y = iris['petal_length'])
#on average, Iris-setosa have least petal_length than others
#mostly whose petal_length are greater belongs to Irsi-virginica
```

Difference between the scatter plot and swarm plot is in case of category to be plotted on  x axis, the vidualization in swarm plot is better than scatter plot because, the data points on the plot does not overlap as it happens in scatter plot leading to better visualization in swarm plot. SO for numeric values scatter plot is good but for categorical values the swarm plot is better

```python
#alloting colors colours to plot graph based on class
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica' : 'b'}
fig, ax = plt.subplots()
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color=colors[iris['class'][i]])
ax.set_title("Iris Dataset")
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_Width')
```

```python
import numpy
fig, ax = plt.subplots()
co = iris["class"].astype('category')
co = co.cat.codes
#because iris['class'] is a object or a kind of category so the above two lines were necessary otherwise if it would have been a numeric value then directly c = iris.iloc[:,2] would have worked 
plt.scatter(iris.iloc[:,0],iris.iloc[:,1],s=100*iris['petal_length'],alpha = 0.3,c = co, cmap='viridis', marker='o')
#marker gives the shape of the datapoints
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
c_bar = plt.colorbar()
c_bar.set_label("class", labelpad = 10)
#labelapd responsible for the distance of label from the colorbar

plt.show()
plt.clf() # clears the plots on the plot stack so that the recent one only comes to picture when show()
```

In the above scatter plot their are 4 properties being represented
First two on axis, third is petal_length which is shown through size varition(s=100*iris['petal_length'])and fourth being the class of iris by color
alpha is used to manage opacity of circles being displayed to ease the visibility of oerlapping circles

Note - if you have created multiple scatter plots (i.e. have enterered plt.scatter() a few times with no call to plt.show()), then these will all be plotted visually on your call to plt.show(). If you are concerned that this is going to happen (and you only want to display your most recent plt.scatter() call), then type plt.clf() which clears everything, and then retype the code to create your figure.

Again, to be safe and ensure everything is clean, type plt.clf() again.

```python
#Line Graph

#removed the column 'class' we dont need to plot it
columns = iris.columns.drop(['class'])

x_data = range(0,iris.shape[0])

fig, ax = plt.subplots()
for column in columns:
    ax.plot(x_data, iris[column], label=column)
           
ax.set_title('Iris Dataset')
ax.legend()
```

```python
cryptoData = pd.read_csv("D:\datasets\crypto-markets.csv",  parse_dates= True)
cryptoData.head(3)
```

```python
filter1 = cryptoData['name']=='Howdoo'
filteredData1 = cryptoData.where(filter1, inplace= False)
filteredData1WithNoNaN = filteredData1.dropna()
a = filteredData1WithNoNaN[['date', 'open']]
a = a.set_index('date')
a = a.rename(columns={'open':'Howdoo_open'})
a
```

```python
filter2 = cryptoData['name'].isin(['VestChain'])
z = cryptoData[filter2]
z = z[['date','open']]
z = z.set_index('date')
z = z.rename(columns={'open':'VestChain_open'})
z
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=a)
sns.lineplot(data=z)
```

```python
aa = pd.concat([a,z],axis=1)
sns.lineplot(data=aa)
```

```python
plt.figure(figsize=(20,6))
plt.title("Crpyto Open")
#no need to for join command below
d = a.join(z, on='date', how = 'outer')
sns.lineplot(data=d)

```
